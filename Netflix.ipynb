```python
plt.figure(figsize=(6,4))
plt.hist(df['release_year'])
plt.xlabel("release_year")
plt.ylabel("Frequency")
```

```python
df['type'].value_counts().plot(kind='bar')
plt.xlabel('type')
plt.ylabel('Frequency')
```

```python
boxes = df["duration_min"].value_counts().head(5)
plt.figure(figsize=(6,6))
plt.pie(boxes.values,
        labels=boxes.index,
        autopct="%1.1f%%",
        startangle=90)
plt.show()
```

```python
df["views_millions"] = pd.to_numeric(df["views_millions"], errors="coerce")
```

```python
plt.figure(figsize=(6,4))

plt.scatter(df["release_year"], df["views_millions"])

plt.title("release_year vs views_millions")
plt.xlabel("release_year")
plt.ylabel("views_millions")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.boxplot(df["duration_min"].dropna())

plt.ylabel("rating")

plt.show()
```

```python
corr = df[["duration_min", "release_year",
           "rating", "views_millions"]].corr()

plt.figure(figsize=(6,5))

plt.imshow(corr, cmap="coolwarm", interpolation="nearest")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.show()
```
